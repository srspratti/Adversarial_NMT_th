```python
import pandas as pd
import json
import requests
```

```python
response_train_all = requests.get("https://datasets-server.huggingface.co/rows?dataset=wmt14&config=fr-en&split=train&offset=0&length=100")
json_response_train_all = response_train_all.json()                      
```

```python
json_response_train_all
```

```python
def get_json_data(json_data):    
    en =list()
    fr = list()
    for row_idx in json_data['rows']:
        en.append(row_idx['row']['translation']['en'])
        fr.append(row_idx['row']['translation']['fr'])
    
    return en, fr
```

```python
en, fr = get_json_data(json_response_train_all)
```

```python
en
```

```python
json_response_train_all = pd.DataFrame(list(zip(en,fr)), columns=['src','target'])
```
